```python
import pandas as pd
cosine_sim_df = pd.read_csv('../dataset/cosine_sim.csv', delimiter=',', encoding='utf-8', index_col='title')
netflix_df = pd.read_csv('../dataset/netflix_titles.csv', delimiter=',', encoding='utf-8')
cosine_sim_df.head()
```

```python
def get_recomendation(title, df=netflix_df, cosine_sim=cosine_sim_df, n=10) -> list:
    try:
        sim = list(enumerate(cosine_sim_df.loc[title]))
        sim = sorted(sim, key= lambda x: x[1], reverse=True)[1:n+1]
        indices = [i[0] for i in sim]
        return df['title'].iloc[indices].to_list()
    except KeyError:
        return 'No so-called film in database'
```
